```python
import pandas as pd
df = pd.read_csv("salaries_by_college_major.csv")
clean_df = df.dropna()
clean_df.head()
#get a specific cell value using a combination of colname and index
clean_df["Undergraduate Major"][clean_df["Mid-Career Median Salary"].idxmax()]
#get the entire row using loc.[index]
clean_df.loc[clean_df["Mid-Career Median Salary"].idxmin()]
clean_df.loc[clean_df["Starting Median Salary"].idxmin()]
#sort values, mutate, and group_by
clean_df[["Undergraduate Major", "Mid-Career Median Salary"]].sort_values("Mid-Career Median Salary", ascending=False)
clean_df.insert(1, "spread", clean_df['Mid-Career 90th Percentile Salary']- clean_df['Mid-Career 10th Percentile Salary'])
clean_df.sort_values("spread")
clean_df.groupby("Group").mean()
#number formatting
pd.options.display.float_format = '{:,.2f}'.format 
```
